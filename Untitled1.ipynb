# Big Data Pipeline: Malicious URL Classification
## 1. System Initialization
In a true Big Data environment, traditional libraries like Pandas cannot handle the volume of data in memory. This section initializes a distributed PySpark Session, configured to utilize all available local CPU cores. We explicitly disable Catalyst constraint propagation to prevent JVM heap exhaustion when processing datasets with high dimensionality (70+ features).

```python
!pip install pyspark -q
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import warnings
warnings.filterwarnings(action='ignore')

from pyspark.sql import SparkSession

# Initialize High-Performance Spark Session
spark = (
    SparkSession.builder
    .appName('CyberThreat_Classifier_BD')
    .master('local[*]')
    .config('spark.driver.memory', '4g')
    .config('spark.sql.constraintPropagation.enabled', 'false')
    .getOrCreate()
)
spark.sparkContext.setLogLevel("OFF")
print("PySpark Distributed Engine Initialized.")
```

## 2. Data Ingestion & Schema Inspection
Here, we ingest the `All.csv` dataset into a Resilient Distributed Dataset (RDD) DataFrame. Before any cleaning occurs, it is critical to inspect the raw schema and data structure to understand the datatypes and feature space we are working with.

```python
file_location = '/content/drive/My Drive/All.csv'
df_raw = spark.read.csv(file_location, header=True, inferSchema=True)

print(f"Total Records Ingested: {df_raw.count()}")
print(f"Total Feature Columns: {len(df_raw.columns)}\n")

print("--- Raw Dataset Schema (Top 10 Features) ---")
# Displaying only the first 10 columns for readability
for field in df_raw.schema.fields[:10]:
    print(f"{field.name}: {field.dataType}")

print("\n--- Raw Data Sample ---")
df_raw.select(df_raw.columns[:5]).show(5)
```

## 3. Data Cleaning & Memory Optimization
Real-world data is inherently messy. We must handle missing values and unparseable strings (like "NaN"). Because PySpark uses lazy evaluation, iterative cleaning over 70+ columns creates a massive logical plan that crashes Colab's RAM.
To bypass this, we use a low-level native type-casting strategy. We then write the clean data to disk (Parquet format) and reload it. This "Checkpointing" severs the execution lineage, freeing up 100% of the memory for analysis and modeling.

```python
from pyspark.sql.functions import col

# Sanitize column names
safe_cols = [c.replace(".", "_") for c in df_raw.columns]
df_working = df_raw.toDF(*safe_cols)

target = "URL_Type_obf_Type"
df_working = df_working.dropna(subset=[target]).where((col(target) != "NaN") & (col(target) != ""))

# Memory-Safe Cast to double
features = [c for c in df_working.columns if c != target]
cast_exprs = [col(c).cast("double").alias(c) for c in features] + [col(target)]
df_cleaned = df_working.select(*cast_exprs).dropna()

# Checkpoint to Parquet to clear JVM Memory
print("Writing intermediate Parquet file to sever Catalyst lineage...")
df_optimized = df_cleaned.limit(4800) # Hardware constraint limit
df_optimized.write.mode("overwrite").parquet("/content/clean_data.parquet")

df_final = spark.read.parquet("/content/clean_data.parquet")
print(f"Cleaned dataset reloaded. Usable records: {df_final.count()}")
```

## 4. Exploratory Data Analysis (EDA)
### 4.1 Target Variable Distribution
Before training a machine learning model, we must identify if our target classes are imbalanced. We perform a distributed `.count()` aggregation in PySpark, and then transition the summarized result to Pandas for visualization using Seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Aggregating target variable frequencies...")
class_distribution = df_final.groupBy(target).count().toPandas()

plt.figure(figsize=(10, 5))
sns.barplot(data=class_distribution, x=target, y="count", palette="viridis")
plt.title("Cyber Threat Class Distribution")
plt.xlabel("Threat Classification")
plt.ylabel("Total Records")
plt.xticks(rotation=45)
plt.show()
```

### 4.2 Feature Statistical Profiling
To understand the numerical scale of our features, we generate distributed summary statistics (Mean, Standard Deviation, Min, Max) for a subset of critical URL attributes. We then aggregate the average values to see how features like URL Length and Entropy vary by threat type.

```python
from pyspark.sql.functions import mean, round as spark_round

print("--- Statistical Summary of Key Features ---")
# Using PySpark's describe() for distributed summary stats
df_final.select("urlLen", "Entropy_URL", "domainlength").describe().show()

print("--- Average Feature Values per Threat Type ---")
feature_summary = df_final.groupBy(target).agg(
    spark_round(mean("urlLen"), 2).alias("Avg_URL_Length"),
    spark_round(mean("Entropy_URL"), 2).alias("Avg_Entropy"),
    spark_round(mean("domainlength"), 2).alias("Avg_Domain_Length")
).toPandas()

display(feature_summary) # Renders a clean table in Colab
```

## 5. Feature Engineering Pipeline
PySpark MLlib requires all independent variables to be unified into a single mathematical vector.
1. **StringIndexer:** Encodes categorical text labels into machine-readable numeric indices.
2. **VectorAssembler:** Condenses the 70+ numerical features into a dense `raw_features` vector.
3. **StandardScaler:** Normalizes the feature vectors to have a mean of 0 and standard deviation of 1, ensuring features with large scales don't dominate the algorithms.

```python
from pyspark.ml.feature import StringIndexer, VectorAssembler, StandardScaler

numeric_features = [
    f.name for f in df_final.schema.fields
    if f.dataType.typeName() in ['integer', 'double'] and f.name != target
]

indexer = StringIndexer(inputCol=target, outputCol="encoded_label", handleInvalid="skip")
assembler = VectorAssembler(inputCols=numeric_features, outputCol="raw_features", handleInvalid="skip")
scaler = StandardScaler(inputCol="raw_features", outputCol="scaled_features", withStd=True, withMean=False)

train_data, test_data = df_final.randomSplit([0.75, 0.25], seed=42)
print("Data partitioned: 75% Training, 25% Testing.")
```

## 6. Distributed Model Training
We define three distinct classification algorithms:
* **Logistic Regression:** Serves as our linear baseline.
* **Decision Tree:** Handles non-linear relationships.
* **Random Forest:** An ensemble method that builds multiple trees to reduce overfitting and capture complex obfuscation patterns.

```python
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression, DecisionTreeClassifier, RandomForestClassifier

print("Compiling ML Pipelines...")
lr = LogisticRegression(featuresCol="scaled_features", labelCol="encoded_label", maxIter=10)
dt = DecisionTreeClassifier(featuresCol="scaled_features", labelCol="encoded_label", maxDepth=5)
rf = RandomForestClassifier(featuresCol="scaled_features", labelCol="encoded_label", numTrees=20, maxDepth=5)

pipe_lr = Pipeline(stages=[indexer, assembler, scaler, lr])
pipe_dt = Pipeline(stages=[indexer, assembler, scaler, dt])
pipe_rf = Pipeline(stages=[indexer, assembler, scaler, rf])

print("Executing Distributed Training...")
model_lr = pipe_lr.fit(train_data)
print(" -> Logistic Regression Trained.")
model_dt = pipe_dt.fit(train_data)
print(" -> Decision Tree Trained.")
model_rf = pipe_rf.fit(train_data)
print(" -> Random Forest Trained.")
```

## 7. Model Evaluation & Conclusion
Because our EDA revealed a class imbalance, Accuracy alone is an insufficient metric. We utilize the **F1-Measure** (the harmonic mean of precision and recall) to evaluate the models against the unseen testing data.

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
import pandas as pd

preds_lr = model_lr.transform(test_data)
preds_dt = model_dt.transform(test_data)
preds_rf = model_rf.transform(test_data)

eval_acc = MulticlassClassificationEvaluator(labelCol="encoded_label", predictionCol="prediction", metricName="accuracy")
eval_f1 = MulticlassClassificationEvaluator(labelCol="encoded_label", predictionCol="prediction", metricName="f1")

results = pd.DataFrame({
    "Model": ["Logistic Regression", "Decision Tree", "Random Forest"],
    "Accuracy": [eval_acc.evaluate(preds_lr), eval_acc.evaluate(preds_dt), eval_acc.evaluate(preds_rf)],
    "F1-Score": [eval_f1.evaluate(preds_lr), eval_f1.evaluate(preds_dt), eval_f1.evaluate(preds_rf)]
})

print("\n--- FINAL PERFORMANCE METRICS ---")
print(results.sort_values(by="Accuracy", ascending=False).to_string(index=False))

# Visualization
df_melted = results.melt(id_vars="Model", var_name="Metric", value_name="Score")
plt.figure(figsize=(9, 5))
sns.barplot(data=df_melted, x="Model", y="Score", hue="Metric", palette="mako")
plt.title("Final Model Evaluation Comparison")
plt.ylim(0.85, 1.0)
plt.show()
```
